```python
import pandas as pd
from ydata_profiling import ProfileReport
import glob
import os
```

```python
csv_files = glob.glob("../data/tables_V2.0/*.csv")

print(f"Đã tìm thấy {len(csv_files)} file CSV. Bắt đầu kiểm tra ...\n")

for file in csv_files:
    print(f"⏳ Đang đọc file: {file}...")
    df = pd.read_csv(file, low_memory=False)
    output_filename = file.replace('.csv', '_Report.html')
    profile = ProfileReport(df, title=f"Báo cáo tình hình - {file}", minimal=True)
    profile.to_file(output_filename)
    print(f"✅ Xong! Đã lưu báo cáo: {output_filename}\n")

print("🎉 TUYỆT VỜI! Đã kiểm tra xong toàn bộ bảng dữ liệu!")
```
